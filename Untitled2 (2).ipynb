```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd
import seaborn as sns
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score
from catboost import CatBoostClassifier
import numpy as np
from sklearn.utils.class_weight import compute_class_weight
```

```python
df = pd.read_parquet('/home/jupyter/datasphere/project/Хахатон/train_dataset_hackaton2023_train (1).gzip')
```

```python
def moving_average(a, n=3):
    ret = np.cumsum(a, dtype=int)
    ret[n:] = ret[n:] - ret[:-n]
    return ret[n - 1:] / n

Q1 =  df['revenue'].quantile(0.25)
Q3 = df['revenue'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - (1.5 * IQR)
upper_bound = Q3 + (1.5 * IQR)

df_new = df.query("@lower_bound < revenue < @upper_bound").reset_index(inplace=False)
df_new = df_new.drop(columns = ['index'], axis = 1)
df_new['date_diff_post'] = df_new['date_diff_post'].fillna(-1)

df_new['date_diff_post'] = df_new['date_diff_post'].fillna(-1)
df_new['day_of_week'] = df_new['startdatetime'].dt.dayofweek
df_new['month'] = df_new['startdatetime'].dt.month
df_new['hour'] = df_new['startdatetime'].dt.hour

gr_y = df_new.groupby(['customer_id', 'startdatetime']).agg({'revenue' : ['mean']}).reset_index()
gr_y.columns = gr_y.columns.droplevel(1)
df_new = df_new.merge(gr_y, on = ['customer_id', 'startdatetime'])
df_new = df_new.rename(columns={'revenue_x': 'revenue', 'revenue_y': 'meanrevenue'})
df_new

d = moving_average(df_new['revenue'].values)
d = np.append(d, [35, 35])
df_new['sk_sr'] = d
```

```python
df_new = df_new.drop(columns = ['date_diff_post','group_name'])
df_new
```

```python
target = 'buy_post'
X = df_new.loc[:, df_new.columns != target]
y = df_new.loc[:, target]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.22, random_state = 2023)
```

```python
from scipy.stats import randint as sp_randint
from sklearn.model_selection import RandomizedSearchCV
features = list(X_train.columns)
cat_features = ['dish_name','format_name']

classes = np.unique(y_test)
weights = compute_class_weight(class_weight='balanced', classes=classes, y=y_test)
class_weights = dict(zip(classes, weights))

model = CatBoostClassifier(task_type="CPU", iterations = 300, random_state = 2023, eval_metric='F1', class_weights = class_weights, learning_rate = 0.3)

model.fit(X,y, cat_features = cat_features, plot = True, eval_set = (X_test,y_test))
```
